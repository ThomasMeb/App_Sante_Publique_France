```python
# Fonction qui trouve les 5 combinaisons d'aliments les plus optimisées sous forme de menu journalier
def find_optimal_food_combinations(nutrient_needs, data):
  # Définition des poids pour chaque groupe PNNS
    pnns_weights = {
      "Sugary snacks": 0.5,
      "Cereals and potatoes": 0.5,
      "Milk and dairy products": 0.8,
      "Composite foods": 0.7,
      "Beverages": 0.5,
      "Fat and sauces": 0.6,
      "Fish Meat Eggs": 0.9,
      "Fruits and vegetables": 0.9,
      "Salty snacks": 0.5
  }
    
    # Création du problème de programmation linéaire
    prob = LpProblem("Menu Optimization", LpMinimize)

    # Définition des variables de décision
    food_vars = LpVariable.dicts("Food", (row["code"] for index, row in data.iterrows()), lowBound=0, cat='Continuous')

    prob += lpSum([food_vars[row["code"]] * (row["energy_100g"] - nutrient_needs["calories"]) for index, row in data.iterrows()]) + \
               lpSum([food_vars[row["code"]] * (row["proteins_100g"] - nutrient_needs["protein"]) for index, row in data.iterrows()]) + \
               lpSum([food_vars[row["code"]] * (row["fat_100g"] - nutrient_needs["fat"]) for index, row in data.iterrows()]) + \
               lpSum([food_vars[row["code"]] * (row["carbohydrates_100g"] - nutrient_needs["carbohydrates"]) for index, row in data.iterrows()]) + \
             lpSum([food_vars[row["code"]] * pnns_weights.get(row["pnns_groups_1"], 0) for index, row in data.iterrows()])

    # Résolution du problème
    solver = GLPK()
    solver.solve(prob)
    
    # Définition des variables de décision
    food_vars = LpVariable.dicts("Food", (index for index, row in data.iterrows()), lowBound=0, cat='Continuous')

    # Affichage des résultats
    print("Voici les 5 combinaisons d'aliments les plus optimisées sous forme de menu journalier :")
    for index, row in data.iterrows():
        if food_vars[f"{index}"].varValue > 0:
            print(f"{row['produit']} ({food_vars[f'{index}'].varValue} portion(s))")  
```
