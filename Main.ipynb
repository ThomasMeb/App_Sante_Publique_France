```python
import pandas as pd
from pulp import *
```

```python
data = pd.read_csv("data_clean.csv")
```

```python
data.loc[data['pnns_groups_1']=='fruits-and-vegetables','pnns_groups_1']='Fruits and vegetables'
data.loc[data['pnns_groups_1']=='sugary-snacks','pnns_groups_1']='Sugary snacks'
data.loc[data['pnns_groups_1']=='cereals-and-potatoes','pnns_groups_1']='Cereals and potatoes'
data.loc[data['pnns_groups_1']=='salty-snacks','pnns_groups_1']='Salty snacks'
```

```python
def select_foods(df, energy, protein, fat, saturated_fat, carbs, sugars):
  # Filtrer le dataframe pour sélectionner les aliments qui appartiennent aux groupes souhaités
  df = df[df['pnns_groups_1'].isin(['Sugary snacks', 'Cereals and potatoes', ...])]
  
  # Trier les aliments en fonction de leur valeur nutritive par portion
  df = df.sort_values(by=['proteins_100g', 'energy_100g'], ascending=[False, True])
  
  # Initialiser un dataframe vide pour stocker les aliments sélectionnés
  selected_foods = pd.DataFrame(columns=df.columns)
  
  # Initialiser les variables de suivi des nutriments sélectionnés
  total_energy = 0
  total_protein = 0
  total_fat = 0
  total_saturated_fat = 0
  total_carbs = 0
  total_sugars = 0
  
  # Pour chaque aliment dans le dataframe trié
  for _, food in df.iterrows():
    # Si les besoins nutritionnels sont satisfaits, arrêter de sélectionner des aliments
    if total_energy >= energy and total_protein >= protein and total_fat >= fat and total_saturated_fat >= saturated_fat and total_carbs >= carbs and total_sugars >= sugars:
      break
      
    # Sinon, ajouter l'aliment au dataframe des aliments sélectionnés
    selected_foods = selected_foods.append(food)
    
    # Mettre à jour les variables de suivi des nutriments sélectionnés
    total_energy += food['energy_100g']
    total_protein += food['proteins_100g']
    total_fat += food['fat_100g']
    total_saturated_fat += food['saturated-fat_100g']
    total_carbs += food['carbohydrates_100g']
    total_sugars += food['sugars_100g']
    
  # Retourner le dataframe des aliments sélectionnés
  return selected_foods

```

```python
# Utiliser la fonction pour générer un menu journalier avec des besoins nutritionnels spécifiques
selected_foods = select_foods(data, 3600, 90, 75, 30, 550, 50)
```

```python
# Fonction qui trouve les 5 combinaisons d'aliments les plus optimisées sous forme de menu journalier
def find_optimal_food_combinations(nutrient_needs, data):
  # Définition des poids pour chaque groupe PNNS
    pnns_weights = {
      "Sugary snacks": 0.5,
      "Cereals and potatoes": 0.5,
      "Milk and dairy products": 0.8,
      "Composite foods": 0.7,
      "Beverages": 0.5,
      "Fat and sauces": 0.6,
      "Fish Meat Eggs": 0.9,
      "Fruits and vegetables": 0.9,
      "Salty snacks": 0.5
  }
    
    # Création du problème de programmation linéaire
    prob = LpProblem("Menu Optimization", LpMinimize)

    # Définition des variables de décision
    food_vars = LpVariable.dicts("Food", (row["product_name"] for index, row in data.iterrows()), lowBound=0, cat='Integer')

    prob += lpSum([food_vars[row["product_name"]] * (row["energy_100g"] - nutrient_needs["calories"]) for index, row in data.iterrows()]) + \
               lpSum([food_vars[row["product_name"]] * (row["proteins_100g"] - nutrient_needs["protein"]) for index, row in data.iterrows()]) + \
               lpSum([food_vars[row["product_name"]] * (row["fat_100g"] - nutrient_needs["fat"]) for index, row in data.iterrows()]) + \
               lpSum([food_vars[row["product_name"]] * (row["carbohydrates_100g"] - nutrient_needs["carbohydrates"]) for index, row in data.iterrows()]) + \
             lpSum([food_vars[row["product_name"]] * pnns_weights[row["pnns_groups_1"]] for index, row in data.iterrows()])

    # Résolution du problème
    prob.solve()

    # Définition des variables de décision
    food_vars = LpVariable.dicts("Food", (index for index, row in data.iterrows()), lowBound=0, cat='Integer')
    # Affichage des résultats
    print("Voici les 5 combinaisons d'aliments les plus optimisées sous forme de menu journalier :")
    for index, row in data.iterrows():
        if food_vars[f"{index}"].varValue > 0:
            print(f"{row['product_name']} ({food_vars[f'{index}'].varValue} portion(s))")  
```
